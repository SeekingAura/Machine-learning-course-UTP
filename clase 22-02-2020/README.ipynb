# Teoria de la probabilidad

La **teoría de la probabilidad** es la herramienta que usamos para cuantificar el nivel de incertidumbre que se tiene sobre el problema de aprendizaje.

# Teoria de la decisión
Posteriormente, la teor´ıa de decision´ usa la teor´ıa de probabilidad para la toma de decisiones.


**retomando** aprnedizaje de maquina es el parendizaje a razón de un conjunto de datos que forma un modelo en donde se le indica a la maquina el como aprender.

D={t<sub>1</sub>, t<sub>2</sub>, t<sub>3</sub>, ..., t<sub>N</sub>} -> Datos
p(w|d) -> Probabilidad de los parametros dado los datos

La probabilidad me permite cuantificar la certidumbre, es decir que tan efectivo es el modelo con los datos, en este caso la idea es maximizar w (parametros del modelo)

max **w** 

p(w|D) = \frac{P(D|w)p(w)}{P(D)}

p(w|D) -> Posterior, son cosas que conozco antes del calculo

p(W) -> Prior, son cosas que conozco despues del calculo

Esto es equivalente a

max P(w|D) = max P(D|w)P(w)

El modelo es
y = f(t|w)

y -> Salida
t -> entrada
w -> Parámetros para alterar o ajustar el modelo
