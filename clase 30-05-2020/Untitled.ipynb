## Algoritmo de las K-medias
* Se busca identificar grupos de datos en un espacio multidimensional
* Un grupo se puede entender como un conjunto de datos cuya distancia entre si es pequeña comparada con la distancia a los puntos por fuera del grupo... es decir la distancia que tiene respecto a los otros datos

* Cada vector se le aplica un prototipo en donde determina la distancia que tiene cada dato respecto a los demas y establecer pertenecia a un grupo

* Encontrar una asignación de los datos observados **x** a los grupos **k** 
* Se define una medida de distorción como la diferencia entre un punto y otro al cuadrado $\sum \sum r_{n,k}||x_{n}-\mu _{k}||^{2}$

donde $r_{n,k}$ es una variable binaria que indica a cual de los K grupos se asigna el vector de observación $x_{n}$ donde vale 1 si pertenece al valor k y si no vale 0.

* El objetivo es encontrar valores $r_{n,k}$ y ${\mu _{k}}$ sea minima.

* Lo anterior se puede lograr mediante un proceso iterativo de dos pasos
    * Se escojen los $\mu_{k}$ y se minimiza J con respecto a los $r_{n,k}$ manteniendo los $\mu_{k}$ fijos
    * Se minimiza J con respecto a los $\mu _{k}$
* si se hace un analisis de derivada se encuentra que
* El estimador de media que se utiliza es a razón de la cantidad de medias de cada uno de los grupos.
* El denominador es igual al número de puntos asignados al grupo k

### Proceso
En el proceso primero se determina los valores $r_{n, k}$ y empezar a justarlos por cada iteración

La linea de clasificación que ha de formarse es una linea recta debido a que estos han de tener una misma covarianza debido a como es la distancia

al tener una clasificación de los parametros aleatorios se establece la pertencnia de lcase y con estos se puede actualizar los valores de variables indicadores buscando el centroide de estos datos.

El proceso se repite hasta que se converga.
