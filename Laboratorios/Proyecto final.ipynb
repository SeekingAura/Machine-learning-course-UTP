```python
from scipy.io import loadmat
# Load datasets
data_reg=loadmat("Carlos Arturo Moreno Tabares - regression.mat")
data_clf=loadmat("Carlos Arturo Moreno Tabares - classification.mat")

print("Cargada")
```

```python
import random
import numpy as np
import joblib
from sklearn.svm import SVC
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, train_test_split
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, MinMaxScaler, normalize


print("  REGRESSION DATA  ".center(100, "%"))

# Regression
X=data_reg["X"]
y=data_reg["y"][0]

# Show shapes
print("  Shapes  ".center(50, "*"))
print("X Shape {}".format(X.shape))
print("y shape {}".format(y.shape))

print("  End of Shapes  ".center(50, "*"))

#{'St_Scaler__copy': True, 'St_Scaler__with_mean': False, 'St_Scaler__with_std': True, 'mdl_reg__cache_size': 2000, 
# 'mdl_reg__gamma': 1.0, 'mdl_reg__kernel': 'linear', 'mdl_reg__max_iter': 30000}

# Prepare grid settings
print("  Grid settings  ".center(50, "#"))
kernel=('linear',)
# kernel=('rbf',)
C=np.linspace(0.1, 1, 5)
gamma=tuple(set(('auto', 'scale') + tuple(np.logspace(-2, 2, 3))))
max_iter=(100000,)
cache_size=(2000,)
# https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
# "accuracy"
# "balanced_accuracy"
# ""
scoring="r2"


# Creacion del pipeline
pipe = Pipeline([('St_Scaler', StandardScaler()),
                 #('MinMax_Scaler', MinMaxScaler()),
                 ('mdl_reg', SVR()),
                 ])


parameters = {'St_Scaler__copy':[True,False], # fijense para los parametros de cada paso del Pipeline se pone primero el "nombreObjetoPipeline__nombreParametro" tantos como quieras
             'St_Scaler__with_mean':[True,False],
             'St_Scaler__with_std':[True,False],
             'mdl_reg__kernel':kernel,
             'mdl_reg__C':C,
             "mdl_reg__gamma":gamma, 
             'mdl_reg__cache_size':cache_size, 
             'mdl_reg__max_iter':max_iter
            }

print("parameters {}".format(parameters))
print("scoring {}".format(scoring))

print("  End of Grid settings  ".center(50, "#"))

# Prepare grid settings
print("  Make Regressions with grid object  ".center(50, "*"))

# Split train and test data (validation data)
Xtrain,Xtest,y_train,y_test = train_test_split(X,y,test_size=0.35)

mdl_reg = GridSearchCV(pipe,parameters,n_jobs=-1, scoring=scoring, verbose=1000)
mdl_reg.fit(Xtrain,y_train)



print("  End of Make Regressions with grid object  ".center(50, "*"))

# Save machine and save sumary
print("  Save 'best' machine for this run  ".center(50, "#"))

print("Best score: {}".format(mdl_reg.best_score_))

print("score from validation data: {}".format(mdl_reg.score(Xtest,y_test)))
print("Best params \n {}".format(mdl_reg.best_params_))

# Maquina con el mejor rank
# clf.best_estimator_
joblib.dump(mdl_reg, '1088029924_reg.sav') # mlp: es la clase con el modelo entrenado, por ustedes. 

df=pd.DataFrame(mdl_reg.cv_results_)
df.to_csv('1088029924_reg.csv')
# df.head(-1)

print("  End of Save 'best' machine for this run  ".center(50, "#"))


print("  END OF REGRESSION DATA  ".center(100, "%"))



```

```python
import random
import numpy as np
import joblib
from sklearn.svm import SVC
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, MinMaxScaler, normalize

print("  CLASIFICATION DATA  ".center(100, "%"))
# Classification
X=data_clf["X"]
y=data_clf["y"][0]

# Show shapes
print("  Shapes  ".center(50, "*"))
print("X Shape {}".format(X.shape))
print("y shape {}".format(y.shape))

print("  End of Shapes  ".center(50, "*"))

# Prepare grid settings
print("  Grid settings  ".center(50, "#"))
kernel=('rbf',)
# kernel=('rbf',)
C=np.linspace(0.1, 1, 3)
gamma=tuple(set(('auto', 'scale') + tuple(np.logspace(-2, 2, 3))))
#gamma=np.logspace(-1, 1, 2)
degree=(2,)
max_iter=(1000,)
cache_size=(2000,)
# https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
# "accuracy"
# "balanced_accuracy"
# ""
scoring="accuracy"
# parameters = {'kernel':kernel, 'C':C, "gamma":gamma, 'max_iter':max_iter, 'cache_size':cache_size, 'degree':degree} #, 'degree':degree}

# Creacion del pipeline
pipe = Pipeline([('St_Scaler',StandardScaler()),
                 #('MinMax_Scaler', MinMaxScaler()),
                 ('Poly', PolynomialFeatures()),
                 
                 ('mdl_clf',SVC()),
                 ])


parameters = {'St_Scaler__copy':[True], # fijense para los parametros de cada paso del Pipeline se pone primero el "nombreObjetoPipeline__nombreParametro" tantos como quieras
             'St_Scaler__with_mean':[True],
             'St_Scaler__with_std':[True],
             'Poly__degree':degree,
             # 'normalize__norm':('l2', 'l1', 'max'),
             # 'normalize__axis':(0, 1),
             'mdl_clf__kernel':kernel,
             "mdl_clf__gamma":gamma, 
             'mdl_clf__C':C,
             'mdl_clf__cache_size':cache_size, 
             'mdl_clf__max_iter':max_iter, 
            }

print("parameters {}".format(parameters))
print("scoring {}".format(scoring))


print("parameters {}".format(parameters))
print("scoring {}".format(scoring))

print("  End of Grid settings  ".center(50, "#"))

# Prepare grid settings
print("  Make Classifications with grid object  ".center(50, "*"))

# Split train and test data (validation data)
Xtrain,Xtest,y_train,y_test = train_test_split(X,y,test_size=0.35)

mdl_clf = GridSearchCV(pipe,parameters,n_jobs=-1, scoring=scoring, verbose=1000)
mdl_clf.fit(Xtrain,y_train)

print("  End of Make Regressions with grid object  ".center(50, "*"))

# Save machine and save sumary
print("  Save 'best' machine for this run  ".center(50, "#"))

print("Best score: {}".format(mdl_clf.best_score_))

print("score from validation data: {}".format(mdl_clf.score(Xtest,y_test)))

print("Best params \n {} ".format(mdl_clf.best_params_))

# Maquina con el mejor rank
# clf.best_estimator_
joblib.dump(mdl_clf, '1088029924_clf.sav') # mlp: es la clase con el modelo entrenado, por ustedes. 

df=pd.DataFrame(mdl_clf.cv_results_)
df.to_csv('1088029924_clf.csv')
# df.head(-1)

print("  End of Save 'best' machine for this run  ".center(50, "#"))


print("  END OF CLASSIFICATION DATA  ".center(100, "%"))
```
